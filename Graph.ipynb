```python
from scholarly import scholarly as sb
import networkx as nx
```

```python
name = "Pieter Abbeel"
```

```python
count = 0
graph = nx.DiGraph()
stack = [name]
seen = []

while len(stack) > 0:

    curr = stack.pop(0)
    seen.append(curr)  # no double counting

    search_query = sb.search_author(curr)
    author = next(search_query).fill(sections=['indices', 'coauthors'])
    
    print(count, 'examining', curr, "at", author.affiliation)

    weight = author.hindex

    for coauthor in author.coauthors:
        print(">> adding", coauthor.name)
        graph.add_weighted_edges_from([(curr, coauthor.name , weight)])  # https://stackoverflow.com/questions/9136539/how-do-weighted-edges-affect-pagerank-in-networkx

        if (coauthor.name not in seen) or (coauthor.name not in stack):
            stack.append(coauthor.name)

    count += 1
    if count > 5:
        break
```

```python
sorted(nx.pagerank(graph), key = nx.pagerank(graph).get, reverse = True)
```
