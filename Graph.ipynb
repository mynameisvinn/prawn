```python
from scholarly import scholarly
import networkx as nx
```

```python
def generate_pg(graph, stack, count):

    seen = []

    while len(stack) > 0:

        curr = stack.pop(0)
        print(count, 'examining', curr)
        seen.append(curr)  # no double counting

        search_query = scholarly.search_author(curr)
        author = next(search_query).fill(sections=['indices', 'coauthors'])

        weight = author.hindex
        
        if weight > 20:
            for coauthor in author.coauthors:
                print(">> adding", coauthor.name)
                graph.add_weighted_edges_from([(curr, coauthor.name , weight)])  # https://stackoverflow.com/questions/9136539/how-do-weighted-edges-affect-pagerank-in-networkx

                if (coauthor.name not in seen) and (coauthor.name not in stack):
                    stack.append(coauthor.name)
        else:
            pass

        count -= 1
        if count == 0:
            break
            
    return graph
```

```python
D = nx.DiGraph()
stack = ['Ken Goldberg']
```

```python
E = generate_pg(D, stack, 500)
```

```python
sorted(nx.pagerank(E), key=nx.pagerank(E).get, reverse=True)
```
